**Tweet Scraping for Sentiment Analysis Without Twitter API**

This web-scraping script uses snscrape to collect tweets of a user inputted query. 

**Import Libraries**

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

**Functions to Scrape and Save Tweets to CSV**

```python
def get_tweets(query, tweets, limit):
    for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    
        # print(vars(tweet))
        # break
        if len(tweets) == limit:
            break
        else:
            tweets.append([tweet.date, tweet.place, tweet.username, tweet.content])
    return tweets
```

```python
def save_to_csv(tweets):
    df = pd.DataFrame(tweets, columns=['Date', 'Location', 'User', 'Tweet'])
    print(df)

    # to save to csv
    df.to_csv('tweets/tweets.csv')
    return tweets
```

```python
query = "(warriors OR celtics) -RT -retweet -like (#warriors OR #celtics OR #nba OR #nbafinals) lang:en until:2022-06-06 since:2022-06-01 -filter:links"
tweets = []
limit = 1000
get_tweets(query, tweets, limit)
save_to_csv(tweets)
```
